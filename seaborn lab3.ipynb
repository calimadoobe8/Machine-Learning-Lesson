```python
import seaborn as sns
import matplotlib.pyplot as plt


data = sns.load_dataset("flights")
data = data.pivot("month", "year", "passengers")

#used to load the "flights" dataset, which is a built-in dataset in seaborn.
# The dataset contains the number of passengers traveling by month and year.
# the pivot method is used to reformat the data into a matrix, with months on the x-axis and years on the y-axis.



sns.heatmap(data = data, annot=True, fmt="d", cmap="YlGnBu")
# creating heatmap of hte data
#annot is used to add numeric annotations to each cell of the heatmap
# fmt="d" parameter specifies the format of the annotations as integers
#rameter sets the color scheme for the heatmap.

plt.show()
```

```python
sns.set_theme(style="white")

df = sns.load_dataset("penguins")

g = sns.JointGrid(data=df, x="body_mass_g", y="bill_depth_mm", space=0)
g.plot_joint(sns.kdeplot,
             fill=True, clip=((432, 4520), (11, 24)),
             thresh=0, levels=100, cmap="rocket")
g.plot_marginals(sns.histplot, color="#03051A", alpha=1, bins=25)
```
